```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np
import tensorflow as tf

import cv2
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
import random

import json

tf.logging.set_verbosity(tf.logging.INFO)
from skimage import data, img_as_float
from skimage.measure import compare_ssim as ssim

import os

from glob import glob

%matplotlib inline
```

```python
our_dir = [y for x in os.walk('/dataset_full/') for y in glob(os.path.join(x[0], '*.png'))]
shr_dir = [y for x in os.walk('/data_oh_rgb/test-images/') for y in glob(os.path.join(x[0], '*.png'))]
```

```python
re_w = 512
re_h = 512

our_data = cv2.imread(our_dir[0])
our_data = cv2.resize(our_data, (re_w, re_h), interpolation=cv2.INTER_CUBIC)
width = 12
height = 12
plt.figure(figsize=(width, height))
plt.axis("off")
plt.imshow(cv2.cvtColor(our_data, cv2.COLOR_BGR2RGB))
plt.show()

shr_data = cv2.imread(shr_dir[0])
shr_data = cv2.resize(shr_data, (re_w, re_h), interpolation=cv2.INTER_CUBIC)
plt.figure(figsize=(width, height))
plt.axis("off")
plt.imshow(cv2.cvtColor(shr_data, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
ssim1 = ssim(our_data, shr_data, multichannel=True)
print(ssim1)
```
