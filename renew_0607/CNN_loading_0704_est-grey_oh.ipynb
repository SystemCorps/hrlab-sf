```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np
import tensorflow as tf

import cv2
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
import random

import json
```

```python
# GPU selection (memory)
import os
#os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"   
#os.environ["CUDA_VISIBLE_DEVICES"]="0"
```

```python
# Data info.

width = 640
height = 480
channel = 1


# resize
r_w = 512
r_h = 512

total_pix = r_w * r_h * channel
```

```python
def data_process(d_dir, split_rate):
    
    num_class = len(d_dir.keys())

    out = {}
    

    for i in range(num_class):
        x_train, x_test = train_test_split(d_dir[i], test_size=split_rate)
        
        if i == 0:
            train_temp = x_train
            test_temp = x_test
        else:
            train_temp.extend(x_train)
            test_temp.extend(x_test)
    
    random.shuffle(train_temp)
    out['train'] = train_temp
    random.shuffle(test_temp)
    out['test'] = test_temp
    
    return out
```

```python
image_dir1 = '/data_oh/test+train/'
class_name1 = ['normal', 'cut']

data_dir1 = {}
for i in range(len(class_name1)):
    temp_dir = image_dir1 + class_name1[i]
    img_files = [[os.path.join(temp_dir, f), np.int16(i)] for f in os.listdir(temp_dir) if f.endswith('.png')]
    data_dir1[i] = img_files
print(len(data_dir1[0]), len(data_dir1[1]))
```

```python
tr_data_dir = {'normal': data_dir1[0], 'cut':data_dir1[1]}
```

```python
ev_data1 = np.zeros((len(tr_data_dir['normal']), total_pix), dtype=np.float32)
ev_label1 = np.zeros((len(tr_data_dir['normal']), 1), dtype=np.int32)

for i in range(len(tr_data_dir['normal'])):
    img = cv2.imread(tr_data_dir['normal'][i][0],-1)
    img2 = cv2.resize(img, (r_w, r_h), interpolation=cv2.INTER_CUBIC)
    ev_data1[i,:] = img2.flatten()
    ev_label1[i] = tr_data_dir['normal'][i][1]
    
ev_data2 = np.zeros((len(tr_data_dir['cut']), total_pix), dtype=np.float32)
ev_label2 = np.zeros((len(tr_data_dir['cut']), 1), dtype=np.int32)

for i in range(len(tr_data_dir['cut'])):
    img = cv2.imread(tr_data_dir['cut'][i][0],-1)
    img2 = cv2.resize(img, (r_w, r_h), interpolation=cv2.INTER_CUBIC)
    ev_data2[i,:] = img2.flatten()
    ev_label2[i] = tr_data_dir['cut'][i][1]
```

```python
ev_data = ev_data2
ev_label = ev_label2

f_dir1 = '/models/Edge/Train_4/model.ckpt-2280.meta'
f_dir2 = '/models/Edge/Train_4/'

saver = tf.train.import_meta_graph(f_dir1)
batch = 32
ev_steps = int(ev_label.shape[0]/batch)

ev_acc = np.zeros(ev_label.shape[0])


with tf.Session() as sess:
    
    new_saver = tf.train.import_meta_graph(f_dir1)
    new_saver.restore(sess, tf.train.latest_checkpoint(f_dir2))
    graph = tf.get_default_graph()
    #for op in tf.get_default_graph().get_operations():
        #print(op.name)
        #print(op.values())
        
    data_in = graph.get_tensor_by_name("fifo_queue_DequeueUpTo:1")

    #feed_dict={data_in:ev_data[0:10], label:ev_label[0:10]}
    keep_prob = graph.get_tensor_by_name("dropout/dropout/keep_prob:0")
    for i in range(ev_steps):
        if i < (ev_steps-1):
            feed_dict={data_in:ev_data[batch*i:batch*(i+1)], keep_prob:1.0}
            label = ev_label[batch*i:batch*(i+1)].reshape(batch)
        else:
            feed_dict={data_in:ev_data[batch*i:], keep_prob:1.0}
            temp = ev_label[batch*i:]
            label = temp.reshape(len(temp))
        
        prob = graph.get_tensor_by_name("softmax_tensor:0")
        classes = graph.get_tensor_by_name("ArgMax:0")
        
        
        result_pr = sess.run(prob, feed_dict=feed_dict)
        result_cl = sess.run(classes, feed_dict=feed_dict)
        
        print(result_pr)
        print(result_cl)
        print(label)

        acc = np.equal(result_cl, label).reshape((1,len(label)))*1
        
        print(acc)
        print(np.mean(acc))
        
        if i < (ev_steps-1):
            ev_acc[batch*i:batch*(i+1)] = acc
        else:
            ev_acc[batch*i:] = acc
```
